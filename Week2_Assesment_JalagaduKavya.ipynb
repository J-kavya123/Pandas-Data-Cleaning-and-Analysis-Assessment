#====Q11===#

```python
import pandas as pd
df = pd.DataFrame({'price': ['100', '200', 'NA', '400']})
df['price'] = pd.to_numeric(df['price'], errors='coerce')
print(df['price'].dtype) 
print(df['price'].sum()) 
```

#===Q13===# The line df_filtered['rating']=df_filtered['rating']*2 may produce a settingWithCopyWarning because df_filtered is a slice of the original dataframe.

```python
import pandas as pd 
df = pd.DataFrame({'rating': [4.5, 3.8, 4.2, 5.0]})
df_filtered = df[df['rating'] > 4] 
df_filtered['rating'] = df_filtered['rating'] * 2   
print(df) 
print(df_filtered) 
```

Expalnation:the warning occurs because pandas is not sure whether df_filtered is a copy or a view of df.as a result the original df is not reliably updated.ti avoid this warning,use.loc[] on the original datframe or create an explicit copy with df_filtered=df[df['rating']>4,copy().

#====Q14===# the lambda funtion returns small for values less than or equals to 2, and the squres of values greater than 2.

```python
import pandas as pd
df = pd.read_csv("indian_startups_fundingDADS.csv")
```

```python
print(df.columns.tolist())
```

```python
df.columns = [
    "startup_name",
    "city",
    "industry",
    "funding_stage",
    "amount",
    "funding_date",
    "investors",
    "founded_year"
]
print(df.columns.tolist())
```

```python
df["startup_name"] = df["startup_name"].str.strip().str.title()
df["city"] = df["city"].str.strip().str.title()
city_map = {
    "Bangalore":"Bengaluru",
    "Blr":"Bengaluru",
    "Bombay":"Mumbai",
    "Bom":"Mumbai",
    "Delhi":"New Delhi",
    "Gurgaon":"Gurugram",
    "Ggn":"Gurugram",
    "Calcutta":"Kolkata",
    "Madras":"Chennai",
    "Hyd":"Hyderabad"
}

df["city"] = df["city"].replace(city_map)
df["amount"] = (
    df["amount"]
    .astype(str)
    .str.replace("₹","",regex=False)
    .str.replace("INR","",regex=False)
    .str.replace("Cr","",regex=False)
    .str.replace("Crore","",regex=False)
    .str.replace(",","",regex=False)
    .str.replace("Undisclosed","",regex=False)
)
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
df["funding_date"] = pd.to_datetime(
    df["funding_date"],
    errors="coerce",
    dayfirst=True
)
df["founded_year"] = pd.to_numeric(
    df["founded_year"],
    errors="coerce"
)
df = df.drop_duplicates()
print(df.head())
```

```python
industry_map = {
    "ecommerce": "Ecommerce",
    "E-Commerce": "Ecommerce",
    "ECOMMERCE": "Ecommerce",

    "EdTech": "EdTech",
    "EDTECH": "EdTech",
    "Ed-Tech": "EdTech",
    "Ed tech": "EdTech",

    "AgriTech": "Agritech",
    "Agri-Tech": "Agritech",
    "Agri Tech": "Agritech",
    "AGRITECH": "Agritech",
    "agritech": "Agritech",

    "Health-Tech": "HealthTech",
    "Health Tech": "HealthTech",
    "healthtech": "HealthTech",
    "HEALTHTECH": "HealthTech",

    "Food Tech": "FoodTech",
    "Food-Tech": "FoodTech",
    "FOOD-TECH": "FoodTech",
    "foodtech": "FoodTech",

    "Fin-Tech": "FinTech",
    "Fintech": "FinTech",
    "FINTECH": "FinTech",

    "AI": "AI/ML",
    "AI ML": "AI/ML",
    "AI-ML": "AI/ML",
    "AIML": "AI/ML"
}

df["industry"] = df["industry"].replace(industry_map)
print(df["industry"].unique())
```

```python
stage_map = {
    "Series-A": "Series A",
    "SERIES A": "Series A",
    "series a": "Series A",
    "series_a": "Series A",

    "Series-B": "Series B",
    "SERIES B": "Series B",
    "series b": "Series B",
    "series_b": "Series B",

    "Series-C": "Series C",
    "SERIES C": "Series C",
    "series c": "Series C",
    "series_c": "Series C",

    "Series-D": "Series D",
    "SERIES D": "Series D",
    "series d": "Series D",

    "Series-E": "Series E",
    "SERIES E": "Series E",
    "series e": "Series E",

    "Seed": "Seed",
    "SEED": "Seed",
    "seed": "Seed",

    "Pre-Seed": "Pre-Seed",
    "Preseed": "Pre-Seed",
    "PRE-SEED": "Pre-Seed",

    "Bridge": "Bridge",
    "bridge": "Bridge",

    "IPO": "IPO",
    "ipo": "IPO",

    "Pre-IPO": "Pre-IPO"
}

df["funding_stage"] = df["funding_stage"].replace(stage_map)
print(df["funding_stage"].unique())
df = df.drop_duplicates()
```

```python
#30
df["funding_tier"] = df["amount"].apply(
    lambda x:
    "Unicorn" if pd.notna(x) and x >= 1000 else
    "Mega" if pd.notna(x) and x >= 500 else
    "Large" if pd.notna(x) and x >= 100 else
    "Mid" if pd.notna(x) and x >= 10 else
    "Small / Unknown"
)
print(df["funding_tier"].value_counts())
```

```python
#34
result = (
    df.groupby("city")
      .agg({
          "startup_name": "count",
          "amount": ["sum", "mean"]
      })
)

result.columns = [
    "startup_count",
    "total_funding",
    "average_funding"
]

result = result.sort_values(
    by="startup_count",
    ascending=False
)

print(result.head(10))
```

```python
#35
result = (
    df.groupby("industry")
      .agg(
          count=("startup_name", "count"),
          average_funding=("amount", "mean")
      )
)

result = result[result["count"] >= 15]

result = result.sort_values(
    by="average_funding",
    ascending=False
)

print(result.head(1))
```
